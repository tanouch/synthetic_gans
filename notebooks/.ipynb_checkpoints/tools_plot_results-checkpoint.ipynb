```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import seaborn as sns

classes = ["N=2", "N=4", "N=9", "N=25"]
locs = ["d=1", "d=3", "d=9", "d=27"]
barWidth = 0.35
r1 = np.arange(len(classes))
```

```python
def plotting_matrix_heatmap(harvest, name):
    classes = ["K=0.5", "K=1", "K=5", "K=20", "K=50"]
    #[5, 15, 50, 200, 1000]
    locs = ["n=5", "n=15", "n=50", "n=200", "n=1000"]
    fig, ax = plt.subplots(figsize=(8.5,8.5))
    im = ax.imshow(harvest, cmap='coolwarm', alpha=0.75)
    ax.set_xticks(np.arange(len(locs)))
    ax.set_yticks(np.arange(len(classes)))
    ax.set_xticklabels(classes)
    ax.set_yticklabels(locs)
    plt.setp(ax.get_xticklabels(), ha="center", rotation_mode="anchor")
    for i in range(len(classes)):
        for j in range(len(locs)):
            text = ax.text(j, i, harvest[i, j], ha="center", va="center", color="k", fontsize=18.5)
    plt.tick_params(axis="both", labelsize=21)
    plt.savefig(name +'.jpeg', bbox_inches="tight")
```
